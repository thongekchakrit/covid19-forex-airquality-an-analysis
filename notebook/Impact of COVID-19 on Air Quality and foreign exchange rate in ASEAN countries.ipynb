Next, the Foreign Exchange Rate was scrapped from Yahoo Finance. To do so, we began by defining a function to scrape through yahoo currency of interest. After the creation of currency_extractor function, a list was created to contain the URLs to all ASEAN currencies. Lastly, 2 loops were created to loop the through the URLs and the currencies was converted into a dataframe.

```python
#creating a function to extract currency and wrangle data
def currency_extractor(x):
    scrapper = BeautifulSoup(x.content, 'lxml') #parsing
    scrape_table = scrapper.find_all('table', 'W(100%)') #selecting table
    df_table = pd.read_html(str(scrape_table)) #transforming to dataframe
    data_forex = df_table[0][['Date', 'Adj Close**']].drop(100) #selecting needed column and dropping texted row
    data_forex.columns = ['Date', 'Value'] #renaming column to date and value
    data_forex['Date'] = pd.to_datetime(data_forex['Date']).dt.strftime('%m-%d-%Y') 
    data_forex['Value'] = pd.to_numeric(data_forex['Value'], errors='coerce') #changing column values to float
    to_datetime(data_forex)
    return(data_forex)
```

```python
#creating a list to contain the urls
#1)Brunei, 2)Cambodia, 3)Indonesia, 4)Laos, 5)Myanmar (Burma), 6)Malaysia, 7)Philippines, 8)Singapore, 9)Thailand, 10)Vietnam
asean = ['https://finance.yahoo.com/quote/BND%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#BN 
         'https://finance.yahoo.com/quote/KHR%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#CM
         'https://finance.yahoo.com/quote/IDR%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#ID
         'https://finance.yahoo.com/quote/USDLAK%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#LA
         'https://finance.yahoo.com/quote/MMK%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#Bur
         'https://finance.yahoo.com/quote/MYR%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#MY
         'https://finance.yahoo.com/quote/PHP%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#PH
         'https://finance.yahoo.com/quote/SGD%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#SG
         'https://finance.yahoo.com/quote/THB%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d',#TH
         'https://finance.yahoo.com/quote/VND%3DX/history?period1=1549065600&period2=1586736000&interval=1d&filter=history&frequency=1d']#VN

#creating a loop to insert and extract currencies
#to make sure the loop is not overloaded, we split the loop into 2 segments
for i in asean:
    brunei_forex = (currency_extractor(insert_url(asean[0])))
    cambodia_forex = (currency_extractor(insert_url(asean[1])))
    indonesia_forex = (currency_extractor(insert_url(asean[2])))
    laos_forex = (currency_extractor(insert_url(asean[3])))
    burma_forex = (currency_extractor(insert_url(asean[4])))
#creating another loop for the remaining currencies
for fx in asean:
    malaysia_forex = (currency_extractor(insert_url(asean[5])))
    philippines_forex = (currency_extractor(insert_url(asean[6])))
    singapore_forex = (currency_extractor(insert_url(asean[7])))
    thailand_forex = (currency_extractor(insert_url(asean[8])))
    vietnam_forex = (currency_extractor(insert_url(asean[9])))
```

The extreme values seem not to be related to air quality measurements, pressure and pol are indicators for to meteorological data. Hence, we decided not to use these indicators for our analysis. 

According to researchers such as Kyrkilis, Chaloulakou, and Kassomenos (2007), AQI is an important (Kyrkilis, Chaloulakou, & Kassomenos, 2007) indicator that represents the overall air quality. AQI is based on six major pollutants: the fine matter particulate matter (PM2.5), ozone (O3), sulfur dioxide (SO2), inhalable particles (PM10), nitrogen dioxides (NO2) and carbon monoxide (CO) (Liu, Li, Yu, & Gu, 2019). However, in this dataset, not all ASEAN members have a record of AQI, hence in addition to AQI, we will analyze PM10 and PM2.5. In the analysis of air quality, the unit used for the pollutants was g/m3.

```python
#filtering for pm10, pm2.5, aqi
air_quality_6c.reset_index(inplace=True)
air_quality_final = air_quality_6c.loc[air_quality_6c['Specie'].isin(['pm10','pm25','aqi'])]
air_quality_final = air_quality_final.loc[air_quality_final['Country'].isin(['Indonesia','Malaysia','Philippines', 'Singapore', 'Thailand', 'China'])]

#finding the aggregated mean of pm10, pm2.5 and aqi
#centralizing values to per country instead of states
air_quality_nostate = air_quality_final.groupby(['Date','Country','Specie']).mean().reset_index()

#spliting the dataset into indivisual country (reshaping)
#reason is that not every country have aqi, pm2.5, pm10
#remove count column
china_air_quality =  air_quality_nostate[air_quality_nostate['Country'] == 'China'].drop(columns='count')
sg_air_quality =  air_quality_nostate[air_quality_nostate['Country'] == 'Singapore'].drop(columns='count')
indo_air_quality =  air_quality_nostate[air_quality_nostate['Country'] == 'Indonesia'].drop(columns='count')
thai_air_quality =  air_quality_nostate[air_quality_nostate['Country'] == 'Thailand'].drop(columns='count')
my_air_quality =  air_quality_nostate[air_quality_nostate['Country'] == 'Malaysia'].drop(columns='count')
ph_air_quality =  air_quality_nostate[air_quality_nostate['Country'] == 'Philippines'].drop(columns='count')
```

__2.3. The impact of COVID-19 on ASEAN foreign exchange rate and air quality__

Here, the relationship between the COVID-19 pandemic, ASEAN foreign exchange, and air quality will be performed. We started by transforming the data into an appropriate shape and following that, we performed a heatmap correlation analysis on our data.

```python
#transforming all COVID-19 data into columns 
covid19_final = covid19_area.drop(columns={'Lat','Long'})
covid19_final = covid19_final.loc[covid19_area['Country'].isin(['Indonesia', 'Malaysia','Philippines', 'Singapore', 'Thailand'])]
covid19_final_all = covid19_final.pivot_table(values=['Confirmed','Recovered', 'Deaths'], index=['Date'], columns='Country')
covid19_final_confirm = covid19_final.pivot_table(values=['Confirmed'], index=['Date'], columns='Country')
covid19_final_all.columns = covid19_final_all.columns.droplevel(0)
covid19_final_confirm.columns = covid19_final_confirm.columns.droplevel(0)
covid19_final_all.reset_index(inplace=True)
covid19_final_confirm.reset_index(inplace=True)
covid19_final_all.columns = ['Date','Indonesia_Confirmed', 'Malaysia_Confirmed', 'Philippines_Confirmed', 'Singapore_Confirmed', 'Thailand_Confirmed',
                        'Indonesia_Recovered', 'Malaysia_Recovered', 'Philippines_Recovered', 'Singapore_Recovered', 'Thailand_Recovered',
                        'Indonesia_Deaths', 'Malaysia_Deaths', 'Philippines_Deaths', 'Singapore_Deaths', 'Thailand_Deaths']

covid19_final_confirm.columns = ['Date','Indonesia_Confirmed', 'Malaysia_Confirmed', 'Philippines_Confirmed', 
                                 'Singapore_Confirmed', 'Thailand_Confirmed']

covid19_final_all.set_index('Date', inplace=True)
covid19_final_confirm.set_index('Date', inplace=True)

#pivoting pollutant indicators to columns and renaming columns to pollutant names
air_quality_aqi_final = air_quality_aqi
air_quality_aqi_final = air_quality_aqi_final.pivot_table(values=['aqi'], index=['Date'], columns='Country')
air_quality_aqi_final.columns = air_quality_aqi_final.columns.droplevel(0) #drop irrelevant multidex
air_quality_aqi_final.reset_index(inplace=True) #drop pollutant column names to level 0
air_quality_aqi_final.columns = ['Date','China_aqi','Malaysia_aqi']
air_quality_aqi_final.set_index('Date', inplace=True)

air_quality_pm10_final = air_quality_pm10
air_quality_pm10_final = air_quality_pm10_final.pivot_table(values=['pm10'], index=['Date'], columns='Country')
air_quality_pm10_final.columns = air_quality_pm10_final.columns.droplevel(0) #drop irrelevant multidex
air_quality_pm10_final.reset_index(inplace=True) #drop pollutant column names to level 0
air_quality_pm10_final.columns = ['Date', 'China_pm10','Indonesia_pm10','Singapore_pm10', 'Thailand_pm10', 'Philippines_pm10']
air_quality_pm10_final.set_index('Date', inplace=True)

air_quality_pm25_final = air_quality_pm25
air_quality_pm25_final = air_quality_pm25_final.pivot_table(values=['pm25'], index=['Date'], columns='Country')
air_quality_pm25_final.columns = air_quality_pm25_final.columns.droplevel(0) #drop irrelevant multidex
air_quality_pm25_final.reset_index(inplace=True) #drop pollutant column names to level 0
air_quality_pm25_final.columns = ['Date', 'China_pm25','Indonesia_pm25','Singapore_pm25', 'Thailand_pm25']
air_quality_pm25_final.set_index('Date', inplace=True)

#scaled forex currencies
forex_asean_f1 = forex_asean_fn
forex_asean_f1.set_index('Date', inplace=True)

#dataframe contain COVID-19 confirmed cases and foreign exchange
#COVID-19 confirmed cases will be used for the analysis
master_sheet = pd.merge(covid19_final_confirm, forex_asean_f1,  left_on='Date', right_index=True, how='inner')
master_sheet.isna().sum() #0 Nan

#dataframe contains COVID-19 confirmed cases and pm10
#repeat the steps above for pm10
master_sheet_pm10 = pd.merge(covid19_final_confirm, air_quality_pm10_final,  left_on='Date', right_index=True, how='inner')
master_sheet_pm10.isna().sum() # 2 Nan for indonesia_pm10 and 1 for sg_pm10
master_sheet_pm10['Indonesia_pm10'] = master_sheet_pm10['Indonesia_pm10'].fillna(master_sheet_pm10['Indonesia_pm10']).mean()
master_sheet_pm10['Singapore_pm10'] = master_sheet_pm10['Singapore_pm10'].fillna(master_sheet_pm10['Singapore_pm10']).mean()


#dataframe contains COVID-19 confirmed cases, pm2.5
#repeat the steps for pm2.5
master_sheet_pm25 = pd.merge(covid19_final_confirm, air_quality_pm25_final,  left_on='Date', right_index=True, how='inner')
master_sheet_pm25.isna().sum() # 0 NaN

#dataframe contains COVID-19 confirmed cases, foreign exchange, pm2.5, AQI
#repeat the steps for AQI
master_sheet_pm25_aqi = pd.merge(master_sheet_pm25, air_quality_aqi_final,  left_on='Date', right_index=True, how='inner')
master_sheet_pm25_aqi.isna().sum() # 19 NaN for china_aqi and 2 for malaysia_aqi
master_sheet_pm25_aqi['China_aqi'] = master_sheet_pm25_aqi['China_aqi'].fillna(master_sheet_pm25_aqi['China_aqi'].mean()) #filling China_aqi column with its mean value
master_sheet_pm25_aqi['Malaysia_aqi'] = master_sheet_pm25_aqi['Malaysia_aqi'].fillna(master_sheet_pm25_aqi['Malaysia_aqi'].mean()) #filling China_aqi column with its mean value
master_sheet_pm25_aqi.head(1) # dataframe has been reduced to 41 entries
```
